```python
# 일단 태국 상위 5개 도시만 크롤링
cities = ['bangkok','phuket','chiang-mai','nonthaburi','surat-thani']# ,'chiang-mai','nonthaburi','surat-thani'
for city in cities : 
    print(str(city))
```

```python
for city in cities : 

    for page in range(5000) : 
        
        try : 
            page = page + 1 # 1부터 시작되게끔 

            # 첫 프론트 페이지 (page1부터)
            front_page_link = 'https://www.thaiyello.com/location/'+str(city)+ '/' + str(page)

            driver.get(front_page_link)

            #  page_source : 현재 활성창의 소스코드를 갖고옴
            req = driver.page_source


            # 프론트 페이지 안에 보이는 모든 기업들의 상세 설명 페이지 링크 갖고오기 
            frontpage_soup = BeautifulSoup(req, 'html.parser')
            div_items = frontpage_soup.find("div", {"id" : "listings"}).find_all('h4')


            main_link = 'https://www.thaiyello.com'

            # hrefs : 프론트 페이지에 있는 모든 상세페이지 링크들
            # [div.find("a")...] 이 코드는 div_items에 있는 것을 div로 뽑고
            # div 안에 있는 a 태그를 찾아서 거기에 있는 href 속성을 뽑아내서 다시 list로 만들라는 뜻
            hrefs = [main_link + div.find("a")['href'] for div in div_items]


            ['현지언어기업명','영문기업명','기업대표전화번호','대표팩스번호','대표이메일','기업홈페이지URL','현지언어기업주소']

            for link in hrefs : 
                try : 

                    driver.get(link)
                    # 현재 활성창의 페이지 갖고오기 
                    req = driver.page_source
                    soup = BeautifulSoup(req, 'html.parser')


                    # 영문담당자명
                    if "Contact Person" in "".join(list(map(lambda x : x.text, soup.find_all('div', {'class' : 'info'})))) : 
                        for i in soup.find_all('div', {'class' : 'info'}) : 
                            if "Contact Person" in i.text :
                                영문담당자명 = i.text.replace('Contact Person','')
                                영문담당자명리스트.append(영문담당자명)
                    else : 
                        영문담당자명리스트.append(np.nan)

                    # 설립일자  (정확히는 설립연도)
                    if "Establishment year" in "".join(list(map(lambda x : x.text, soup.find_all('div', {'class' : 'info'})))) : 
                        for i in soup.find_all('div', {'class' : 'info'}) : 
                            if "Establishment year" in i.text :
                                설립일자 = i.text.replace('Establishment year','')
                                설립일자리스트.append(설립일자)
                    else : 
                        설립일자리스트.append(np.nan)

                    try : 
                        # 기업대표전화번호 
                        기업대표전화번호 = soup.find("div", {'class' : 'text phone'}).text
                        기업대표전화번호리스트.append(기업대표전화번호)
                    except : 
                        기업대표전화번호리스트.append(np.nan)
                        pass

                    try : 
                        # 영문기업상세주소
                        영문기업상세주소=soup.find("div", {'class' : 'text location'}).text.replace('View Map', '')
                        영문기업상세주소리스트.append(영문기업상세주소)
                    except : 
                        영문기업상세주소리스트.append(np.nan)
                        pass

                    try : 
                        # 영문기업명 
                        영문기업명 = soup.find("b", {'id' : 'company_name'}).text
                        영문기업명리스트.append(영문기업명)
                    except : 
                        영문기업명리스트.append(np.nan)
                        pass

                    try : 
                        # 기업홈페이지URL
                        기업홈페이지URL = soup.find("div", {'class' : 'text weblinks'}).text
                        기업홈페이지URL리스트.append(기업홈페이지URL)
                    except : 
                        기업홈페이지URL리스트.append(np.nan)
                        pass

                    try : 
                        # 영문산업군명
                        영문산업군명 = soup.find("div", {'class' : 'categories'}).text.replace('\n','')
                        영문산업군명리스트.append(영문산업군명)
                    except : 
                        영문산업군명리스트.append(np.nan)
                        pass

                    try : 
                        # 영문기업소개내용
                        영문기업소개내용 = soup.find("div", {'class' : 'text desc'}).text[:400]
                        영문기업소개내용리스트.append(영문기업소개내용)
                    except : 
                        영문기업소개내용리스트.append(np.nan)
                        pass

                except : 
                    continue

            # 페이지가 150이 될때마다 저장. 약 기업 
            if page%150 == 0 :
                save_name = 'Thailand_company_list_'+str(city)+ str(page)+'.xlsx'

                result_df = pd.DataFrame(columns = ['키값', '헤브론스타국가코드', '현지언어국가명', '영문국가명', '국가언어코드값', '현지언어언어명', '영문언어명',
                                                    '현지언어기업명', '영문기업명', '현지언어한줄소개내용', '현지언어기업소개내용', '영문한줄소개내용', '영문기업소개내용',
                                                    '현지언어대표이사명', '영문대표이사명', '기업대표전화번호', '대표팩스번호', '대표이메일', '기업홈페이지URL',
                                                    '페이스북URL', '인스타그램URL', '유튜브URL', '링크드인URL', '현지언어담당자명', '영문담당자명','현지언어직위명', 
                                                    '영문직위명', '담당자부서명', '기업담당자전화번호', '담당자이메일', '담당자팩스번호',
                                                    '기업담당자이동전화번호', '현지언어기업주소', '현지언어기업상세주소', '영문기업주소', '영문기업상세주소', '기업우편번호',
                                                    '현지언어산업군명', '영문산업군명', '현지언어주요제품내용', '영문주요제품내용', '설립일자', '기업종업원값',
                                                    '외감법인구분값', '매출년도값', '매출액값', '기본화폐키값', '화폐단위명', '주식시장코드', '현지언어주식시장명',
                                                    '영문주식시장명', '상장코드', '상장일자', '기업추가정보값', '작업구분코드', '데이터생성일자', '연계처리상태코드',
                                                    '연계처리일자'])

                result_df['영문담당자명'] = 영문담당자명리스트
                result_df['설립일자'] = 설립일자리스트
                result_df['기업대표전화번호'] = 기업대표전화번호리스트
                result_df['영문기업상세주소'] = 영문기업상세주소리스트
                result_df['영문기업명'] = 영문기업명리스트
                result_df['기업홈페이지URL'] = 기업홈페이지URL리스트
                result_df['영문산업군명'] =  영문산업군명리스트
                result_df['영문기업소개내용'] = 영문기업소개내용리스트

                result_df.to_excel(path + save_name, index=False) 

                # reset 
                영문담당자명리스트 = []
                설립일자리스트 = []
                기업대표전화번호리스트 = []
                영문기업상세주소리스트= []
                영문기업명리스트= []
                기업홈페이지URL리스트= []
                영문산업군명리스트= []
                영문기업소개내용리스트= []

            # 페이지가 끝이나서 에러가 났을 경우, 여태까지 긁어 온 정보를 저장하고 루프를 탈출 
        except : 
            save_name = 'Thailand_company_list_'+str(city)+ str(page)+'.xlsx'
            
            result_df = pd.DataFrame(columns = ['키값', '헤브론스타국가코드', '현지언어국가명', '영문국가명', '국가언어코드값', '현지언어언어명', '영문언어명',
                                                '현지언어기업명', '영문기업명', '현지언어한줄소개내용', '현지언어기업소개내용', '영문한줄소개내용', '영문기업소개내용',
                                                '현지언어대표이사명', '영문대표이사명', '기업대표전화번호', '대표팩스번호', '대표이메일', '기업홈페이지URL',
                                                '페이스북URL', '인스타그램URL', '유튜브URL', '링크드인URL', '현지언어담당자명', '영문담당자명','현지언어직위명', 
                                                '영문직위명', '담당자부서명', '기업담당자전화번호', '담당자이메일', '담당자팩스번호',
                                                '기업담당자이동전화번호', '현지언어기업주소', '현지언어기업상세주소', '영문기업주소', '영문기업상세주소', '기업우편번호',
                                                '현지언어산업군명', '영문산업군명', '현지언어주요제품내용', '영문주요제품내용', '설립일자', '기업종업원값',
                                                '외감법인구분값', '매출년도값', '매출액값', '기본화폐키값', '화폐단위명', '주식시장코드', '현지언어주식시장명',
                                                '영문주식시장명', '상장코드', '상장일자', '기업추가정보값', '작업구분코드', '데이터생성일자', '연계처리상태코드',
                                                '연계처리일자'])
            result_df['영문담당자명'] = 영문담당자명리스트
            result_df['설립일자'] = 설립일자리스트
            result_df['기업대표전화번호'] = 기업대표전화번호리스트
            result_df['영문기업상세주소'] = 영문기업상세주소리스트
            result_df['영문기업명'] = 영문기업명리스트
            result_df['기업홈페이지URL'] = 기업홈페이지URL리스트
            result_df['영문산업군명'] =  영문산업군명리스트
            result_df['영문기업소개내용'] = 영문기업소개내용리스트

            result_df.to_excel(path + save_name, index=False, engine='xlsxwriter') 

            # 해당 도시 크롤링을 탈출하고 다음 도시 정보로 이동. 
            continue
            
    
```

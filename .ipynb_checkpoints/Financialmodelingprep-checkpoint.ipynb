### 1-1. 모든 Symbol Listup 

FMP에서는 Fianancial Statements를 갖고 있는 Symbol을 추출하는 API를 제공한다.     
  
ex) https://financialmodelingprep.com/api/v3/financial-statement-symbol-lists?apikey=YOUR_API_KEY 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
# Financial Statments가 있는 기업들의 Symbol을 가져오기 위한 모듈 설치 
try:
    # For Python 3.0 and later
    from urllib.request import urlopen
except ImportError:
    # Fall back to Python 2's urllib2
    from urllib2 import urlopen

import certifi
```

아래 get_jsonparsed_data는 는 FMP document에서 갖고 온 코드로서, 재무제표가 있는 Symbol을 추출할 때 필요한 함수이다. 

```python
def get_jsonparsed_data(url):
    """
    Receive the content of ``url``, parse it as JSON and return the object.

    Parameters
    ----------
    url : str

    Returns
    -------
    dict
    """
    response = urlopen(url, cafile=certifi.where())
    data = response.read().decode("utf-8")
    return json.loads(data)
```

Financial Statments가 있는 기업들의 Symbol을 가져오기 위한 QueryString.

```python
# Financial Statments가 있는 기업들의 Symbol을 가져오기 위한 QueryString. API Key = 33d67d9c5e5ed94ff134e9bf93f9d818
url = ("https://financialmodelingprep.com/api/v3/financial-statement-symbol-lists?apikey=33d67d9c5e5ed94ff134e9bf93f9d818")
```

```python
Symbols = get_jsonparsed_data(url)
```

# 2. 나라마다 재무제표 추출

API를 통해 원하는 재무정보를 갖고 오기 위해는 FMP의 international Filings(해외기업 재무제표) API의 구조를 알아야 한다.  
FMP API는 아래의 Query String에 APIKey를 기입하여 requests.get()으로 손쉽게 데이터를 가져올 수 있는 구조이다.  
API는 json 형태로 데이터를 반환한다. 

ex) https://financialmodelingprep.com/api/v3/income-statement/RY.TO?limit=10&apikey=YOUR_API_KEY 

API Query String 분석 : 
* RY.TO : 요청하고자 하는 상장사 Symbol. 예시로 든 RY.TO는 Toronto(캐나다)에 상장된, Symbol이 RY인 회사이다. 
* limit=   : limit 뒤에 있는 숫자로 현재로부터 과거 몇년간의 데이터를 가져올지를 결정한다. limit=3은 과거 3년간의 데이터 를 갖고오게 됌. 
* apikey=  : 구입한 API key를 기입한다. 

한 나라에 대해 재무제표를 추출하고 전처리 하는 과정을 정해두면, 다른 나라에 대해서도 코드를 재활용할 수 있다.  
프랑스를 기준으로 데이터 수집 및 전처리 코드를 먼저 작성한 후 이를 전체 국가로 확장해본다. 

### 2-1. 프랑스 상장사들의 재무제표 갖고오기 

API는 json 형태로 데이터를 반환한다. 아래 extractor 함수는 URL을 입력하면 API를 통해 반환받은 json을 pandas dataframe으로 반환하는 함수이다.

```python
def extractor(url) : 
    try:
        from pandas import json_normalize
    except ImportError:
        from pandas.io.json import json_normalize
    # 데이터 갖고오기
    req=requests.get(url)
    # json load
    data = json.loads(req.text)
    # json to pandas
    preprocessed = json_normalize(data)
    return preprocessed
```

아래 url_generator 함수는 세가지 인풋을 입력받는다. 
* 재무제표를 수집하려는 기업의 Symbol
* 재무제표 타입(filing_type) : is(income-statement), bs(balance-sheet), cf(cash-flow) 
* 현재부터 과거 몇년치까지 수집할 것인지 제한 : limit  
인풋 변수를 입력하면 각각에 맞는 url을 반환하는 함수이다.

```python
def url_generator(target_Symbol, filing_type, limit) : 
    base_url = 'https://financialmodelingprep.com/api/v3/income-statement/RY.TO?limit=5&apikey=89d4891348727c3950b79b9067127c3f'
    is_url = base_url.replace('RY.TO', target_Symbol)
    is_url = is_url.replace('=5', '='+str(limit))
    bs_url = is_url.replace('income-statement','balance-sheet-statement')
    cf_url = is_url.replace('income-statement','cash-flow-statement')
    
    if filing_type == 'is' : 
        return is_url
    elif filing_type == 'bs' : 
        return bs_url
    elif filing_type == 'cf' : 
        return cf_url 
    else :
        return "Please input proper filing_type"
```

FMP API에서 갖고 올 수 있는 계정과목들은 많다. 그러나 2022년 BDC과제에서 작성한 테이블 명세서에 근거하여 추출하고자 하는 계정과목이름만 리스트에 저장한다. 이렇게 만들어진 리스트들로 필요한 열들만 추출한다. 

```python
bs_cols = ['fillingDate','totalCurrentAssets','totalNonCurrentAssets','totalAssets', 'totalCurrentLiabilities','totalNonCurrentLiabilities',
           'totalLiabilities','totalEquity','totalLiabilitiesAndTotalEquity']
is_cols = ['revenue','costOfRevenue','sellingGeneralAndAdministrativeExpenses','operatingIncome','interestIncome','interestExpense',
          'incomeBeforeTax','netIncome','operatingIncomeRatio'] 
cf_cols = ['netCashProvidedByOperatingActivities','netCashUsedForInvestingActivites','netCashUsedProvidedByFinancingActivities',
          'cashAtBeginningOfPeriod','cashAtEndOfPeriod']
```

```python
result_df = pd.DataFrame() 



for target_Symbol in France_symbols[2:3] :  
    
    # 기업별 URL 생성 
    is_url = url_generator(target_Symbol, 'is', 3)
    bs_url = url_generator(target_Symbol, 'bs', 3)
    cf_url = url_generator(target_Symbol, 'cf', 3)
    
    # extractor 함수를 이용하여 API를 통한 데이터를 판다스 데이터프레임으로 갖고옴. 
    df_is = extractor(is_url)
    df_bs = extractor(bs_url)
    df_cf = extractor(cf_url)
    
    # merge above three dataframes 
    company_df = pd.merge(df_is, df_bs)
    company_df = pd.merge(company_df, df_cf)
    
    # 필요한 열들만 
    company_df = company_df[bs_cols+is_cols+cf_cols]
```

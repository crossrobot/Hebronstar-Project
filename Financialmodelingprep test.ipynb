```python
import requests
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
def extractor(url) : 
    try:
        from pandas import json_normalize
    except ImportError:
        from pandas.io.json import json_normalize
    # 데이터 갖고오기
    req=requests.get(url)
    # json load
    data = json.loads(req.text)
    # json to pandas
    preprocessed = json_normalize(data)
    return preprocessed
```

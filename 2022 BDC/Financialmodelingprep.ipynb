```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
# Financial Statments가 있는 기업들의 Symbol을 가져오기 위한 모듈 설치 
try:
    # For Python 3.0 and later
    from urllib.request import urlopen
except ImportError:
    # Fall back to Python 2's urllib2
    from urllib2 import urlopen

import certifi
from tqdm import tqdm
```

아래 get_jsonparsed_data는 는 FMP document에서 갖고 온 코드로서, 재무제표가 있는 Symbol을 추출할 때 필요한 함수이다. 

```python
def get_jsonparsed_data(url):
    """
    Receive the content of ``url``, parse it as JSON and return the object.

    Parameters
    ----------
    url : str

    Returns
    -------
    dict
    """
    response = urlopen(url, cafile=certifi.where())
    data = response.read().decode("utf-8")
    return json.loads(data)
```

### 1-2. Symbol들을 국가별로 분류한다.

# 2. 나라마다 재무제표 추출

API를 통해 원하는 재무정보를 갖고 오기 위해는 FMP의 international Filings(해외기업 재무제표) API의 구조를 알아야 한다.  
FMP API는 아래의 Query String에 APIKey를 기입하여 requests.get()으로 손쉽게 데이터를 가져올 수 있는 구조이다.  
API는 json 형태로 데이터를 반환한다. 

ex) https://financialmodelingprep.com/api/v3/income-statement/RY.TO?limit=10&apikey=YOUR_API_KEY 

API Query String 분석 : 
* RY.TO : 요청하고자 하는 상장사 Symbol. 예시로 든 RY.TO는 Toronto(캐나다)에 상장된, Symbol이 RY인 회사이다. 
* limit=   : limit 뒤에 있는 숫자로 현재로부터 과거 몇년간의 데이터를 가져올지를 결정한다. limit=3은 과거 3년간의 데이터를 갖고오게 됌. 
* apikey=  : 구입한 API key를 기입한다. 

한 나라에 대해 재무제표를 추출하고 전처리 하는 과정을 정해두면, 다른 나라에 대해서도 코드를 재활용할 수 있다.  
프랑스를 기준으로 데이터 수집 및 전처리 코드를 먼저 작성한 후 이를 전체 국가로 확장해본다. 

### 2-1. 프랑스 상장사들의 재무제표 갖고오기 

API는 json 형태로 데이터를 반환한다. 아래 extractor 함수는 URL을 입력하면 API를 통해 반환받은 json을 pandas dataframe으로 반환하는 함수이다.

```python
def extractor(url) : 
    try:
        from pandas import json_normalize
    except ImportError:
        from pandas.io.json import json_normalize
    # 데이터 갖고오기
    req=requests.get(url)
    # json load
    data = json.loads(req.text)
    # json to pandas
    preprocessed = json_normalize(data)
    return preprocessed
```

아래 url_generator 함수는 세가지 인풋을 입력받는다. 
* 재무제표를 수집하려는 기업의 Symbol
* 재무제표 타입(filing_type) : is(income-statement), bs(balance-sheet), cf(cash-flow) 
* 현재부터 과거 몇년치까지 수집할 것인지 제한 : limit  
인풋 변수를 입력하면 각각에 맞는 url을 반환하는 함수이다.
* 연간 혹은 분기 선택(period) : 'annual' or 'quarter'

```python
def url_generator(target_Symbol, filing_type, limit,period='quarter') : 
    base_url = 'https://financialmodelingprep.com/api/v3/income-statement/RY.TO?limit=5&period=quarter&apikey=89d4891348727c3950b79b9067127c3f'
    if period == 'annual' : 
        base_url = base_url.replace('&period=quarter','')
    
    is_url = base_url.replace('RY.TO', target_Symbol)
    is_url = is_url.replace('=5', '='+str(limit))
    bs_url = is_url.replace('income-statement','balance-sheet-statement')
    cf_url = is_url.replace('income-statement','cash-flow-statement')
    
    if filing_type == 'is' : 
        return is_url
    elif filing_type == 'bs' : 
        return bs_url
    elif filing_type == 'cf' : 
        return cf_url 
    else :
        return "Please input proper filing_type"
```

FMP API에서 갖고 올 수 있는 계정과목들은 많다. 그러나 2022년 BDC과제에서 작성한 테이블 명세서에 근거하여 추출하고자 하는 계정과목이름만 리스트에 저장한다. 이렇게 만들어진 리스트들로 필요한 열들만 추출할 것이다. 

```python
# 재무정보에 반드시 있어야 핵심정보 
# link와 finalLink는 API의 원천 데이터 소스 링크
# isActivelyTrading, isEtf, isAdr, isFund 열들로 수집후 데이터 전처리 과정에서 
# 펀드나 etf가 아닌 현재 상장되있는 기업의 재무제표만 골라내야 
core_cols = ['date','symbol','reportedCurrency','calendarYear','period','link','finalLink']

# 손익계산서에서 추출할 핵심정보 
is_cols = ['revenue','costOfRevenue','sellingGeneralAndAdministrativeExpenses','operatingIncome','totalOtherIncomeExpensesNet',
          'incomeBeforeTax','incomeTaxExpense','netIncome','operatingIncomeRatio'] 
# 대차대조표에서 추출할 핵심정보 
bs_cols = ['totalCurrentAssets','totalNonCurrentAssets','totalAssets', 'totalCurrentLiabilities','totalNonCurrentLiabilities',
           'totalLiabilities','totalEquity','totalLiabilitiesAndTotalEquity']
# 현금흐름표에서 추출할 핵심정보 
cf_cols = ['netCashProvidedByOperatingActivities','netCashUsedForInvestingActivites','netCashUsedProvidedByFinancingActivities',
          'cashAtBeginningOfPeriod','cashAtEndOfPeriod']
```

##### Caution
현재 url generator의 함수의 period변수로 'annual'을 부여함. 우리는 분기별 재무정보를 갖고와야 하기 때문에 period 변수에 'querter'를 부여해야함. 그러나 period 변수에 'quarter'를 부여할 경우 아래 코드를 돌리게 되면 에러가 발생함. 
오류 회피 코드를 만들어, 분기별 재무제표로 뽑았을 때 에러가 나오는 기업의 경우 연간 재무제표를 뽑을 수 있도록 해야할 것 같다. 

```python
company_df_list = pd.DataFrame()
for target_Symbol in tqdm(US_symbols) :  
    
    for report_type in ['annual','quarter'] :
        
        if report_type == 'annual' : 
            limit = 6 # 연간 재무제표 5개년 
        else : 
            limit = 24 # 분기 별 재무제표 (1년에 4분기 * 5 + 여분)
        
    
        # 기업별 URL 생성 
        is_url = url_generator(target_Symbol, 'is', limit, report_type)
        bs_url = url_generator(target_Symbol, 'bs', limit, report_type)
        cf_url = url_generator(target_Symbol, 'cf', limit, report_type)

        # extractor 함수를 이용하여 API를 통한 데이터를 판다스 데이터프레임으로 갖고옴. 
        df_is = extractor(is_url)
        df_bs = extractor(bs_url)
        df_cf = extractor(cf_url)


        # 오류 회피 
        # 간혹가다 손익 계산서, 대차대조표는 있는데 현금흐름표가 없는등의 문제가 있다. 
        # 모든 경우에 대한 오류 회피 
        # try : 만약 추출해 올 열들에 대한 정보가 존재한다면 불러온다
        # except : 만약 추출 해 올 열들에 대한 정보가 있으면, core_is()

        try :  
            df_is = df_is[core_cols+is_cols]
        except : 
            df_is = pd.DataFrame(columns=core_cols+is_cols)    
        try :     
            df_bs= df_bs[core_cols+bs_cols]
        except : 
            df_bs = pd.DataFrame(columns=core_cols+bs_cols)

        try : 
            df_cf = df_cf[core_cols+cf_cols]
        except : 
            df_cf = pd.DataFrame(columns=core_cols+cf_cols)


        # merge above three dataframes 
        #company_df = pd.concat([df_is,df_bs,df_cf],axis = 1)
        company_df = pd.merge(df_is, df_bs, how = 'outer', on = core_cols)
        company_df = pd.merge(company_df, df_cf, how = 'outer', on = core_cols)

        # 숫자 데이터 이외에 필요한 기업 일반정보들을 추가한다. 
        # Financil Modeling Prep의 Company profile API를 이용한다. 
        base_url = 'https://financialmodelingprep.com/api/v3/profile/RY.TO?apikey=89d4891348727c3950b79b9067127c3f'
        target_url = base_url.replace('RY.TO', target_Symbol)

        # request로 데이터 갖고오기 
        req = requests.get(target_url)
        data = json.loads(req.text)

        # 데이터 명세서에 근거하여 필요한 일반정보 칼럼들만 추출하여 company_df에 추가한다. 

        try : company_df['companyName'] = data[0]['companyName'] 
        except :company_df['companyName'] = ""
        try : company_df['ceo'] = data[0]['ceo']
        except :company_df['ceo'] = ""
        try : company_df['phone'] =data[0]['phone']
        except : company_df['phone'] = ""
        try : company_df['website'] =data[0]['website']
        except : company_df['website'] = ""
        try : company_df['state'] =data[0]['state']
        except : company_df['state'] = ""
        try : company_df['city'] =data[0]['city']
        except : company_df['city'] = ""
        try : company_df['country'] =data[0]['country']
        except : company_df['country'] = ""
        try : company_df['industry'] =data[0]['industry']
        except : company_df['industry'] = ""
        try : company_df['ipoDate'] =data[0]['ipoDate']
        except : company_df['ipoDate'] = ""
        try : company_df['address'] =data[0]['address']
        except : company_df['address'] = ""
        try : company_df['zip'] =data[0]['zip']
        except : company_df['zip'] = ""
        try : company_df['exchangeShortName'] =data[0]['exchangeShortName']
        except : company_df['exchangeShortName'] = ""
        try : company_df['exchange'] =data[0]['exchange']
        except :  company_df['exchange'] = ""
        try : company_df['description'] =data[0]['description']
        except : company_df['description'] = ""
        try : company_df['isEtf'] =data[0]['isEtf']
        except : company_df['isEtf'] = ""
        try : company_df['isActivelyTrading'] =data[0]['isActivelyTrading']
        except : company_df['isActivelyTrading'] = ""
        try : company_df['isFund'] =data[0]['isFund']
        except : company_df['isFund'] = ""
            
        # 택스 번호 추가. 'Company core information API'를 이용한다. 
        try : 
            base_url = 'https://financialmodelingprep.com/api/v4/company-core-information?symbol=AAPL&apikey=89d4891348727c3950b79b9067127c3f'
            target_url = base_url.replace('AAPL', target_Symbol)

            # request로 데이터 갖고오기 
            # 갖고오는 데이터가 null이게 되면, 'data'에서 정보를 추출하는 과정에서 에러가 발생하게 될 것이다. 
            req = requests.get(target_url)
            data = json.loads(req.text)
            
            company_df['taxIdentificationNumber'] =data[0]['taxIdentificationNumber']
            company_df['registrantName'] =data[0]['registrantName']

        except : 
            company_df['taxIdentificationNumber'] = ""
            company_df['registrantName'] = ""
        
            
            
        company_df_list = company_df_list.append(company_df)
        
    
    
company_df_list = company_df_list.sort_values(by=['symbol','date'], ascending = [True,False]) # symbol은 오름차순으로, 같은 심볼 내에서 회계분기는 내림차순이 되도록 정렬.        

# 2022년 현재, 5개년치의 데이터, 즉 2017년 이상인 데이터만 뽑아낸다. -> calendarYear(회계연도)가 2017년 이상 
company_df_list = company_df_list[company_df_list['calendarYear'].astype('int')>=2017]
company_df_list.to_excel('US_Original_220609.xlsx', index=False)
```

```python
class FMP_extractor : 
    def __init__(self, apikey, mapping_sheet_dir) : 
        # Financial Statments가 있는 기업들의 Symbol을 가져오기 위한 QueryString. API Key = 33d67d9c5e5ed94ff134e9bf93f9d818
        self.apikey = apikey
        self.url = "https://financialmodelingprep.com/api/v3/financial-statement-symbol-lists?apikey=33d67d9c5e5ed94ff134e9bf93f9d818"
        self.Symbols = self.get_jsonparsed_data(self.url)
        # 재무정보에 반드시 있어야 핵심정보 
        # link와 finalLink는 API의 원천 데이터 소스 링크
        # isActivelyTrading, isEtf, isAdr, isFund 열들로 수집후 데이터 전처리 과정에서 
        # 펀드나 etf가 아닌 현재 상장되있는 기업의 재무제표만 골라내야 
        self.core_cols = ['date','symbol','reportedCurrency','calendarYear','period','link','finalLink']

        # 손익계산서에서 추출할 핵심정보 
        self.is_cols = ['revenue','costOfRevenue','sellingGeneralAndAdministrativeExpenses','operatingIncome','totalOtherIncomeExpensesNet',
                  'incomeBeforeTax','incomeTaxExpense','netIncome','operatingIncomeRatio'] 
        # 대차대조표에서 추출할 핵심정보 
        self.bs_cols = ['totalCurrentAssets','totalNonCurrentAssets','totalAssets', 'totalCurrentLiabilities','totalNonCurrentLiabilities',
                   'totalLiabilities','totalEquity','totalLiabilitiesAndTotalEquity']
        # 현금흐름표에서 추출할 핵심정보 
        self.cf_cols = ['netCashProvidedByOperatingActivities','netCashUsedForInvestingActivites','netCashUsedProvidedByFinancingActivities',
                  'cashAtBeginningOfPeriod','cashAtEndOfPeriod']
        
        ### 1-2. Symbol들을 국가별로 분류한다. 
        self.US_symbols = []          # 
        self.Canada_symbols = []      #TO
        self.France_symbols = []      #PA ? 
        self.Germany_symbols = []     #DE
        self.India_symbols = []       #NS
        self.London_symbols = []      #L
        self.Hongkong_symbols = []    #HK
        self.Austrailia_symbols = []  #AX
        self.Swiss_symbols = []       #SW
        self.Korea_symbols = []       #KR
        self.Netherlands_symbols = [] #EU ? 
        self.symbol_classifier()
        
        
        ## 대응표 갖고오기 
        self.mapping_sheet = pd.read_excel('FMP_API계정과목대응표.xlsx')
        

    # ------------------------ method is below -------------------# 
    def symbol_classifier(self) :
        for Symbol in self.Symbols : 
            if ".TO" in Symbol : 
                self.Canada_symbols.append(Symbol) 
            elif ".PA" in Symbol : 
                self.France_symbols.append(Symbol) 
            elif ".DE" in Symbol : 
                self.Germany_symbols.append(Symbol) 
            elif ".NS" in Symbol : 
                self.India_symbols.append(Symbol) 
            elif ".BSE" in Symbol : 
                self.India_symbols.append(Symbol) 
            elif ".L" in Symbol : 
                self.London_symbols.append(Symbol) 
            elif ".HK" in Symbol : 
                self.Hongkong_symbols.append(Symbol) 
            elif ".AX" in Symbol : 
                self.Austrailia_symbols.append(Symbol) 
            elif ".SW" in Symbol : 
                self.Swiss_symbols.append(Symbol) 
            elif "." not in Symbol :
                self.US_symbols.append(Symbol) 
                        
        return
   
        
    def get_jsonparsed_data(self,url):
        """
        Receive the content of ``url``, parse it as JSON and return the object.

        Parameters
        ----------
        url : str

        Returns
        -------
        dict
        """
        response = urlopen(url, cafile=certifi.where())
        data = response.read().decode("utf-8")
        return json.loads(data)


    def extractor(self,url) :
        #extractor 함수는 URL을 입력하면 API를 통해 반환받은 json을 pandas dataframe으로 반환하는 함수
        try:
            from pandas import json_normalize
        except ImportError:
            from pandas.io.json import json_normalize
            
            
        # 데이터 갖고오기
        req=requests.get(url)
        # json load
        data = json.loads(req.text)
        # json to pandas
        preprocessed = json_normalize(data)
        return preprocessed

    def url_generator(self, target_Symbol, filing_type, limit,period='quarter') : 
        """
        rl_generator 함수는 세가지 인풋을 입력받는다.

        1.재무제표를 수집하려는 기업의 Symbol
        2.재무제표 타입(filing_type) : is(income-statement), bs(balance-sheet), cf(cash-flow)
        3.현재부터 과거 몇년치까지 수집할 것인지 제한 : limit

        인풋 변수를 입력하면 각각에 맞는 url을 반환하는 함수이다.
        연간 혹은 분기 선택(period) : 'annual' or 'quarter'
        """
        base_url = 'https://financialmodelingprep.com/api/v3/income-statement/RY.TO?limit=5&period=quarter&apikey=89d4891348727c3950b79b9067127c3f'
        if period == 'annual' : 
            base_url = base_url.replace('&period=quarter','')

        is_url = base_url.replace('RY.TO', target_Symbol)
        is_url = is_url.replace('=5', '='+str(limit))
        bs_url = is_url.replace('income-statement','balance-sheet-statement')
        cf_url = is_url.replace('income-statement','cash-flow-statement')

        if filing_type == 'is' : 
            return is_url
        elif filing_type == 'bs' : 
            return bs_url
        elif filing_type == 'cf' : 
            return cf_url 
        else :
            return "Please input proper filing_type"

        
    def extract_and_save(self, nation, bgn_year, end_year ,save_dir) : 
        if nation == 'US' : symbols = self.US_symbols
        elif nation == 'Canada' : symbols = self.Canada_symbols
        elif nation == 'France' : symbols = self.France_symbols
        elif nation == 'Germany' : symbols = self.Germany_symbols
        elif nation == 'India' : symbols = self.India_symbols
        elif nation == 'London' : symbols = self.London_symbols
        elif nation == 'Hongkong' : symbols = self.Hongkong_symbols
        elif nation == 'Austrailia' : symbols = self.Austrailia_symbols
        elif nation == 'Swiss' : symbols = self.Swiss_symbols
        
  
        company_df_list = pd.DataFrame()
        for target_Symbol in tqdm(symbols) :  

            for report_type in ['annual','quarter'] :

                if report_type == 'annual' : 
                    limit = end_year-bgn_year # 연간 재무제표 몇년치를 획득할지 결정 
                else : 
                    limit = (end_year-bgn_year)*5 + 3 # 분기 별 재무제표를 몇년치 횔득할지 결정 (1년에 4분기 * 5 + 여분)


                # 기업별 URL 생성 
                is_url = self.url_generator(target_Symbol, 'is', limit, report_type)
                bs_url = self.url_generator(target_Symbol, 'bs', limit, report_type)
                cf_url = self.url_generator(target_Symbol, 'cf', limit, report_type)

                # extractor 함수를 이용하여 API를 통한 데이터를 판다스 데이터프레임으로 갖고옴. 
                df_is = self.extractor(is_url)
                df_bs = self.extractor(bs_url)
                df_cf = self.extractor(cf_url)


                # 오류 회피 
                # 간혹가다 손익 계산서, 대차대조표는 있는데 현금흐름표가 없는등의 문제가 있다. 
                # 모든 경우에 대한 오류 회피 
                # try : 만약 추출해 올 열들에 대한 정보가 존재한다면 불러온다
                # except : 만약 추출 해 올 열들에 대한 정보가 있으면, core_is()

                try :  
                    df_is = df_is[self.core_cols+self.is_cols]
                except : 
                    df_is = pd.DataFrame(columns=self.core_cols+self.is_cols)    
                try :     
                    df_bs= df_bs[self.core_cols+self.bs_cols]
                except : 
                    df_bs = pd.DataFrame(columns=self.core_cols+self.bs_cols)

                try : 
                    df_cf = df_cf[self.core_cols+self.cf_cols]
                except : 
                    df_cf = pd.DataFrame(columns=self.core_cols+self.cf_cols)


                # merge above three dataframes 
                #company_df = pd.concat([df_is,df_bs,df_cf],axis = 1)
                company_df = pd.merge(df_is, df_bs, how = 'outer', on = self.core_cols)
                company_df = pd.merge(company_df, df_cf, how = 'outer', on = self.core_cols)

                # 숫자 데이터 이외에 필요한 기업 일반정보들을 추가한다. 
                # Financil Modeling Prep의 Company profile API를 이용한다. 
                base_url = 'https://financialmodelingprep.com/api/v3/profile/RY.TO?apikey=89d4891348727c3950b79b9067127c3f'
                target_url = base_url.replace('RY.TO', target_Symbol)

                # request로 데이터 갖고오기 
                req = requests.get(target_url)
                data = json.loads(req.text)

                # 데이터 명세서에 근거하여 필요한 일반정보 칼럼들만 추출하여 company_df에 추가한다. 

                try : company_df['companyName'] = data[0]['companyName'] 
                except :company_df['companyName'] = ""
                try : company_df['ceo'] = data[0]['ceo']
                except :company_df['ceo'] = ""
                try : company_df['phone'] =data[0]['phone']
                except : company_df['phone'] = ""
                try : company_df['website'] =data[0]['website']
                except : company_df['website'] = ""
                try : company_df['state'] =data[0]['state']
                except : company_df['state'] = ""
                try : company_df['city'] =data[0]['city']
                except : company_df['city'] = ""
                try : company_df['country'] =data[0]['country']
                except : company_df['country'] = ""
                try : company_df['industry'] =data[0]['industry']
                except : company_df['industry'] = ""
                try : company_df['ipoDate'] =data[0]['ipoDate']
                except : company_df['ipoDate'] = ""
                try : company_df['address'] =data[0]['address']
                except : company_df['address'] = ""
                try : company_df['zip'] =data[0]['zip']
                except : company_df['zip'] = ""
                try : company_df['exchangeShortName'] =data[0]['exchangeShortName']
                except : company_df['exchangeShortName'] = ""
                try : company_df['exchange'] =data[0]['exchange']
                except :  company_df['exchange'] = ""
                try : company_df['description'] =data[0]['description']
                except : company_df['description'] = ""
                try : company_df['isEtf'] =data[0]['isEtf']
                except : company_df['isEtf'] = ""
                try : company_df['isActivelyTrading'] =data[0]['isActivelyTrading']
                except : company_df['isActivelyTrading'] = ""
                try : company_df['isFund'] =data[0]['isFund']
                except : company_df['isFund'] = ""

                # 택스 번호 추가. 'Company core information API'를 이용한다. 
                try : 
                    base_url = 'https://financialmodelingprep.com/api/v4/company-core-information?symbol=AAPL&apikey=89d4891348727c3950b79b9067127c3f'
                    target_url = base_url.replace('AAPL', target_Symbol)

                    # request로 데이터 갖고오기 
                    # 갖고오는 데이터가 null이게 되면, 'data'에서 정보를 추출하는 과정에서 에러가 발생하게 될 것이다. 
                    req = requests.get(target_url)
                    data = json.loads(req.text)

                    company_df['taxIdentificationNumber'] =data[0]['taxIdentificationNumber']
                    company_df['registrantName'] =data[0]['registrantName']

                except : 
                    company_df['taxIdentificationNumber'] = ""
                    company_df['registrantName'] = ""

                company_df_list = company_df_list.append(company_df)



        company_df_list = company_df_list.sort_values(by=['symbol','date'], ascending = [True,False]) # symbol은 오름차순으로, 같은 심볼 내에서 회계분기는 내림차순이 되도록 정렬.        

        # bgn_year 이상인 데이터만 뽑아낸다. -> calendarYear(회계연도)가 bgn_year 이상인 제무제표들
        # 회계 연도가 na인 행들이 있다. 이런 행들은 제거한다. 결측치를 제거하지 않으면 오류가 난다. 
        company_df_list = company_df_list.dropna(subset=['calendarYear'])
        
        company_df_list = company_df_list[company_df_list['calendarYear'].astype('int')>=bgn_year]
        company_df_list.to_csv(save_dir, index=False)
        
        return company_df_list
    
    def preprocess(self, df, save_dir) : 
        # 대응표를 딕셔너리화 key : value = 기준필드명 : FMP필드명
        mapping_dic = self.mapping_sheet.set_index('채워야할 테이블 필드명').T.to_dict('index')['Financial Modeling API']
        processed = pd.DataFrame(columns = mapping_dic.keys())
        for order, field in enumerate(processed.columns) : 
            try : 
                processed.loc[:,field] = df.loc[:,mapping_dic[field]]
            except : 
                pass
        import re 
        regExp = '\W[a-zA-Z]+'  # \W로 문자, 숫자가 아닌 특수문자 제거하고 [a-zA-Z]+ 로 ''. 뒤에있는 주식시장 코드 제거
        processed['상장코드']=processed['상장코드'].str.replace(pat = regExp ,repl=r'', regex = True)

        processed.to_csv(save_dir, index = False)
        
        return
```

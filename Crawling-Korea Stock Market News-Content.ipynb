```python
import pandas as pd 
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
from selenium.common.exceptions import NoSuchElementException 
import time 
import re 
from tqdm import tqdm 
import selenium
import bs4
from selenium import webdriver
from bs4 import BeautifulSoup
import html5lib
import requests
import math
import numpy as np
```

```python
def company_content(company_df) : 
    
    path = "C:\\Users\\user\\chromedriver.exe"
    driver = webdriver.Chrome(path)
    # 크롬 웹 드라이버 객체 생성.
    # 크롬드라이버로 크롬을 조작.

    driver.get("https://finance.naver.com")
    #드라이버 객체에서 원하는 url로 이동.

    company_df['이벤트 내용'] = np.nan

    for i in range(len(company_df)) : 
        url = company_df['이벤트URL'][i]
        try : 
            driver.get(url)
            page = driver.page_source 
            soup = BeautifulSoup(page,'html.parser') 
            fullcontent = soup.find('div', attrs = {'class':'articleCont'}).text
            summary = '. '.join(fullcontent.strip().split('.')[0:3])

            # 문장 길이가 4000자 이상(띄어쓰기 포함)이면 summary의 길이를 잘라준다. 
            if len(summary) > 4000 : 
                summary = summary[0:4000] 


            company_df.loc[i,'이벤트 내용']=summary
        
        except : 
            next
        
    return company_df
```
